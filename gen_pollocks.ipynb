```python
import pickle
import pandas as pd

# Load new_df.pkl
path = 'new_df.pkl'
with open(path, 'rb') as f:
    new_df = pickle.load(f)
new_df.head()
```

```python
print(len(new_df))
```

```python
# List of elem in new_df['quantity']
quantity_list = new_df['quantity'].unique().tolist()

# List of elem in new_df['exposure']
exposure_list = new_df['exposure'].unique().tolist()

# List of elem in new_df['c7_labels']
cluster_list = new_df['c7_labels'].unique().tolist()

print(quantity_list)
print(exposure_list)
print(cluster_list)
```

```python
import matplotlib.pyplot as plt

# For each elem in cluster_list, quantity_list, and exposure_list, plot the trajectories
for q in quantity_list:
    for c in cluster_list:
        for e in exposure_list:
            title = 'c_' + str(c) + '-' + 'e_' + str(e) + '-' + 'q_' + str(q)
            print(title)
            for index, row in new_df.iterrows():
                if (row['exposure'] == e) & (row['c7_labels'] == c) & (row['quantity'] == q):        
                    x = row['traj'][0]
                    y = row['traj'][1]
                    plt.plot(x, y)
            plt.title(title)
            filename = 'img/' + title + '.png'
            plt.savefig(filename) 
            plt.close("all")
```
