```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load new_df.pkl
path = 'new_df.pkl'
with open(path, 'rb') as f:
    new_df = pickle.load(f)
new_df.sample(10)
```

```python
# Select 20 random samples from new_df which c7_values are greater than 0.7 and c7_labes are 5
sample_df = new_df[(new_df['c7_values'] > 0.7) & (new_df['c7_labels'] == 5)].sample(20)
sample_df.head(20)
```

```python
# Store the values of sample_df['traj] into a list
traj_list = sample_df['traj'].tolist()
traj_list[0]
```

```python
import numpy as np

def zero_origin_traj(traj):
    """
    This function is used to change the origin of the trajectory to (0, 0)
    """
    # Get the first point of the trajectory
    x = traj[0]
    y = traj[1]
    x -= x[0]
    y -= y[0]
    new_traj = np.array([x, y])

    return new_traj
```

```python

zo_traj = zero_origin_traj(traj_list[0])
zo_traj
```

```python
# Plot the trajectory
plt.plot(zo_traj[0], zo_traj[1], 'r-')
```

```python
# Plot the trajectories in traj_list
for traj in traj_list:
    zo_traj = zero_origin_traj(traj)
    plt.plot(zo_traj[0], zo_traj[1])
plt.show()
```
